```python
#architecture of lstm
#layers and shape
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import tensorflow as tf  #mathmatical function,tensors manuplating tools
from tensorflow.keras import layers 
#keras is an api for tensorflow jaha operation kr skte hai

vocab_size=1000 #unique words or unique tokens
embedding_dim=16
sequence_length=8

#model architecture of lstm

lstm_model = Sequential(
    [
                #layer 1 ;input wording embeddings
                Embedding(
                    input_dim=vocab_size,
                    output_dim=embedding_dim,
                    input_length=sequence_length, # will takes 8 words , 8 integers
                    name="embedding_layer"
                ),

                #layer 2: Sequence processing (RNN base-with lstm touch)
                LSTM(units=32,name="lstm_layer"), # it will process 32 internal memory cell info --- it procesess 8 vectors

                #layer 3:output binary classification 
                Dense(units=1,activation="sigmoid",name="output_layer")



    ],
     name="Sequential_LSTM_Classifier"

)
#3 verification 
lstm_model.summary()
```
